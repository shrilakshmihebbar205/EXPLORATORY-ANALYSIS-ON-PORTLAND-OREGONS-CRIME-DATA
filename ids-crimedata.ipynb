```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import datetime
df=pd.read_csv("dataset.csv")
df
```

```python
##inserting colons inbetween time
df["Occur Time"]=df["Occur Time"].apply(lambda x:'{0:0>4}'.format(x))
df["Occur Time"]=df["Occur Time"].astype(str)
df["Occur Time"]=df["Occur Time"].str[:2]+':'+df["Occur Time"].str[-2:]
df["Occur Time"]
```

```python
##extracting hour from given time
df['hour'] = pd.to_datetime(df["Occur Time"],format='%H:%M').dt.hour
df['hour']
```

```python
##remove 3% null values from neighborhood
df['Neighborhood'].replace('',np.nan, inplace=True)
df.dropna(subset=['Neighborhood'],inplace=True)
df.Neighborhood
```

```python
##remove 10% null values from address
df['Address'].replace('',np.nan, inplace=True)
df.dropna(subset=['Address'],inplace=True)
df.Address
```

```python
df['OpenDataX'].replace('',np.nan, inplace=True)
df.dropna(subset=['OpenDataX'],inplace=True)
df.OpenDataX
```

```python
df['OpenDataY'].replace('',np.nan, inplace=True)
df.dropna(subset=['OpenDataY'],inplace=True)
df.OpenDataY
```

```python
##bring all dates to same format
li=[]
for index,row in df["Occur Date"].iteritems():
    row=row.replace('-','/')
    li.append(row)
df=pd.DataFrame({'col':li})
print(df)
```

```python
##bring all dates to same format
li=[]
for index,row in df["Report Date"].iteritems():
    row=row.replace('-','/')
    li.append(row)
df1=pd.DataFrame({'col':li})
print(df1)
```

```python
##extracting year and month from occur month year
df['year'] = pd.DatetimeIndex(df["Occur Month Year"]).year
df['month'] = pd.DatetimeIndex(df["Occur Month Year"]).month
df['year']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point,Polygon

%matplotlib inline
df=pd.read_csv("dataset.csv")
crs={'init':'epsg:4326'}
df.head()
```

```python
geometry = [Point(xy) for xy in zip( df["OpenDataLon"],df["OpenDataLat"])]
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(df, crs= crs, geometry=geometry)
geo_df.head()
```

```python
street_map=gpd.read_file('oregon_administrative.shp')
```

```python
fig,ax = plt.subplots(figsize=(15,15))
street_map.plot(ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
street_map.plot(ax=ax,alpha=0.4,color="grey")
geo_df[geo_df['OpenDataX']>700000].plot(ax=ax, markersize=20, color="blue" , marker="o",label="Neg")
geo_df[geo_df['OpenDataY']<700000].plot(ax=ax, markersize=20, color="red" , marker="^",label="Pos")
plt.legend(prop={'size':15})
```
